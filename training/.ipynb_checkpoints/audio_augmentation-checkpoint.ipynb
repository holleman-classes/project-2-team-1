**Time and frequency**

`time_stretch(y, *, rate, **kwargs)`

Time-stretch an audio series by a fixed rate.

`pitch_shift(y, *, sr, n_steps[, ...])`

Shift the pitch of a waveform by n_steps steps.

```python
!ls ./custom-speech/computer | head -n 5
```

```python
fs, wav = wavfile.read('./custom-speech/computer/computer_171206.wav')
wav = wav / np.max(np.abs(wav))
```

```python
plt.plot(np.arange(len(wav))/fs, wav)
ipd.Audio(wav, rate=fs)
```

https://librosa.org/doc/latest/effects.html

```python
wav_long = lreff.time_stretch(wav, rate=0.6) # rate>1 => faster/shorter
```

```python
print(len(wav))
print(len(wav_long))
```

```python
plt.plot(np.arange(len(wav_long))/fs, wav_long)
ipd.Audio(wav_long, rate=fs)
```

```python
wav_up = lreff.pitch_shift(wav, sr=fs, n_steps=6, bins_per_octave=12) 
wav_dn = lreff.pitch_shift(wav, sr=fs, n_steps=-3, bins_per_octave=12) 
```

```python
plt.plot(np.arange(len(wav_up))/fs, wav_up)
ipd.Audio(wav_up, rate=fs)
```

```python
plt.plot(np.arange(len(wav_dn))/fs, wav_dn)
ipd.Audio(wav_dn, rate=fs)
```

```python
wav_trim, idx = lreff.trim(wav, top_db=40)
```

```python
plt.plot(np.arange(len(wav_trim))/fs, wav_trim)
```
